# Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#In a firm of 400 employees, the percentage of monthly salary
#saved by each employee is given in the following table. Represent it through a bar graph.
#-----------------------------------------------------------------
#		Savings (in percentage)	Number of Employees(Frequency)
#------------------------------------------------------------------
#			20					105
#			30					199
#			40					29
#			50					73
```

```python
save = np.array([20,30,40,50])
n_emp = np.array([105,199,29,73])
print("Savings(%)\tNo.of empolyee")
print('----------------------')
for sv,ep in zip(save,n_emp):
    print("{}\t\t{}".format(sv,ep))
```

```python
plt.bar(save,n_emp)
plt.show()
```

```python
plt.bar(save,n_emp,width=4,color="purple")
plt.show()
```

```python
plt.bar(save,n_emp,width=4,color="purple")
plt.xlabel("Savings (%)")
plt.ylabel("No. of Employee")
plt.title("Employees and their savings (vertical graph)")
plt.show()
```

```python
# Displying the horizontal graph
plt.barh(save,n_emp,height=4,color='orange')
plt.show()
```

```python
plt.barh(save,n_emp,height=4,color='orange')
plt.xlabel('Savings (%)')
plt.ylabel("No.of Employee")
plt.title("Employees and their savings (horizontal graph)")
plt.show()
```

```python
# Draw a bar chart for year and number of students placed	(Vertical bar and Horizontal )					

#		Years			Number of students placed
#		---------			-------------------------------------
#		2020				120
#		2021				200
#		2022				175
#		2023				100
```

```python
data = {'Year':[2020,2021,2022,2023],'No. of stud':[120,200,175,100]}
print('------------------------')
for yr,stud in data.items():
    print("\t{}".format(yr))
    for n in stud:
        print("\t\t{}".format(n))
```

```python
# converting dict data to dataframe object
df = pd.DataFrame(data)
print(df)

# now plot into bar graph
plt.bar(df['Year'],df['No. of stud'],color='purple')
plt.show()
```

```python
plt.bar(df['Year'],df['No. of stud'],color='hotpink')
plt.xlabel("Years")
plt.ylabel("No. of students")
plt.title("Number of students are placed")
plt.show()
```

```python
plt.barh(df['Year'],df['No. of stud'],color='green')
plt.xlabel("No of students")
plt.ylabel("Years")
plt.title("No of students are placed (Horizontal graph)")
plt.show()
```

```python
# Example of bar Graph
x = ["Apples","Kiwi","Sbrerry","Mango"]
y = [40,20,35,23]
colors = ["red","brown","hotpink","yellow"]
plt.bar(x,y,color=colors)
plt.show()
```

```python
x = ["Apples","Kiwi","Sbrerry","Mango"]
y = [40,20,35,23]
colors = ["red","brown","hotpink","yellow"]
plt.bar(x,y,color=colors,width=0.5)
plt.show()
```

```python
x = ["Apples","Kiwi","Sbrerry","Mango"]
y = [40,20,35,23]
colors = ["red","brown","hotpink","yellow"]
lab = ['Apples','kiwi','sbrerry','Mongo']
plt.bar(x,y,color=colors,width=0.5,label=lab)
plt.legend()
plt.show()
```

```python
# Display the same example of graph in horizontal
x = ["Apples","Kiwi","Sbrerry","Mango"]
y = [40,20,35,23]
colors = ["red","brown","hotpink","yellow"]
lab = x
plt.barh(x,y,color=colors,height=0.5,label=lab)
plt.legend()
plt.show()
```

```python
x = ["Apples","Kiwi","Sbrerry","Mango"]
y = [40,20,35,23]
plt.xlabel("Y-Values")
plt.ylabel("X-Values")
plt.title("Example of Bar Graph")
plt.barh(x,y,color=['yellow','pink','purple','red'],label=['Apples','kiwi','Sbrerry','Mongo'],height=0.5)
plt.legend()
plt.show()
```
