```python
# Legend property in Graph
import matplotlib.pyplot as plt
import numpy as np
a=np.array([1,2,3,4,5])
b=a**2
c=a**3
plt.plot(a,a,ls="solid",lw=6,c="r",marker="o",mec="r",mfc="g",ms=5)
plt.plot(a,b,ls="solid",lw=6,c="k",marker="d",mec="g",mfc="r",ms=6)
plt.plot(a,c,ls="solid",lw=6,c="hotpink",marker="+",mec="k",mfc="r",ms=7)
#Font for Labels of x,y and title
f1={"font" : "Arial", "size":12,"color":"red"}
f2={"font" : "Arial", "size":18,"color":"blue"}
plt.xlabel("X-Axis",fontdict=f1)
plt.ylabel("Y-Axis",fontdict=f1)
plt.title("Natural Numbers, Square and Cubes",fontdict=f2)
lab = ['Natural number','squares','cubes']
plt.legend(labels=lab,facecolor='lightblue',edgecolor='b',fontsize=12)
plt.grid(axis="both",ls="dotted",lw=2,c="y") 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.array([1,2,3,4,5])
b=a**2
c=a**3
plt.plot(a,a,ls="solid",lw=6,c="r",marker="o",mec="r",mfc="g",ms=5)
plt.plot(a,b,ls="solid",lw=6,c="k",marker="d",mec="g",mfc="r",ms=6)
plt.plot(a,c,ls="solid",lw=6,c="hotpink",marker="+",mec="k",mfc="r",ms=7)
#Font for Labels of x,y and title
f1={"font" : "Arial", "size":12,"color":"red"}
f2={"font" : "Arial", "size":18,"color":"blue"}
plt.xlabel("X-Axis",fontdict=f1)
plt.ylabel("Y-Axis",fontdict=f1)
plt.title("Natural Numbers, Square and Cubes",fontdict=f2)
lab = ['Natural number','squares','cubes']
plt.legend(labels=lab,loc='lower center',facecolor='lightblue',edgecolor='b',fontsize=12)
plt.grid(axis="both",ls="dotted",lw=2,c="y") 
plt.show()
```
