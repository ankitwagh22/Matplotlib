# Markers Concept in line graphs

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b)
plt.show()
```

```python
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b,marker="o")
plt.show()
```

```python
# ms --> marker size
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b,marker="p",ms=10)
plt.show()
```

```python
# mec = marker edge color
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b,marker="p",ms=10,mec="r")
plt.show()
```

```python
# mfc = marker face color
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b,marker="p",ms=10,mec="b",mfc='r')
plt.show()
```

```python
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b,marker="p",ms=10,mec="b",mfc='r',ls="dashed")
plt.show()
```

```python
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b,marker="p",ms=10,mec="b",mfc='r',ls="dotted")
plt.show()
```

```python
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b,marker="p",ms=10,mec="b",mfc='r',ls="dashdot")
plt.show()
```

```python
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b,marker="p",ms=10,mec="b",mfc='r',ls="None")
plt.show()
```

```python
# lw  -------> line width
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b,marker="p",ms=10,mec="b",mfc='r',ls="solid",lw=5)
plt.show()
```

```python
#  c--------->Line color
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
plt.plot(a,b,marker="p",ms=10,mec="b",mfc='r',ls="solid",lw=5,c='y')
plt.show()
```

```python
# Problem---Draw a Line Graph for For Natural Numbers, squares and Cubes of Natural Numbers
import matplotlib.pyplot as plt
import numpy as np
arr = np.array([1,2,3,4,5])
squ = arr ** 2
cube = arr ** 3
plt.plot(arr,arr,marker='o',ms=9,mfc='b',mec='r',lw=5)
plt.plot(arr,squ,marker='s',ms=9,mfc='b',mec='r',lw=3)
plt.plot(arr,cube,marker='+',ms=8,mfc='b',mec='r',lw=5)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Natural Numbers Squares & Cubes")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
arr = np.array([1,2,3,4,5])
squ = arr ** 2
cube = arr ** 3
plt.plot(arr,arr,marker='o',ms=9,mfc='b',mec='r',lw=5)
plt.plot(arr,squ,marker='s',ms=9,mfc='b',mec='r',lw=3)
plt.plot(arr,cube,marker='+',ms=8,mfc='b',mec='r',lw=5)

# Font for Labels of x,y and title
f1 = {'font':'Arial','size':12,'color':'red'}
plt.xlabel("X-Axis",fontdict=f1)
plt.ylabel("Y-Axis",fontdict=f1)
plt.title("Natural Numbers Squares & Cubes",fontdict=f1)
plt.show()
```

```python
# Grid Properties
import matplotlib.pyplot as plt
import numpy as np
a=np.array([2,3,4,5])
b=np.array([1,2,3,5])
plt.plot(a,b,ls="-",lw=8,c="hotpink",marker="p",mec="g",mfc="y",ms=15)
plt.grid() # it is defult grid
plt.show()
```

```python
# Grid Properties ---> x-axis
import matplotlib.pyplot as plt
import numpy as np
a=np.array([2,3,4,5])
b=np.array([1,2,3,5])
plt.plot(a,b,ls="-",lw=8,c="hotpink",marker="p",mec="g",mfc="y",ms=15)
plt.grid(axis='x') 
plt.show()
```

```python
# Grid Properties ---> y-axis
import matplotlib.pyplot as plt
import numpy as np
a=np.array([2,3,4,5])
b=np.array([1,2,3,5])
plt.plot(a,b,ls="-",lw=8,c="hotpink",marker="p",mec="g",mfc="y",ms=15)
plt.grid(axis='y') 
plt.show()
```

```python
# Grid Properties ---> both
import matplotlib.pyplot as plt
import numpy as np
a=np.array([2,3,4,5])
b=np.array([1,2,3,5])
plt.plot(a,b,ls="-",lw=8,c="hotpink",marker="p",mec="g",mfc="y",ms=15)
plt.grid(axis='both') 
plt.show()
```
