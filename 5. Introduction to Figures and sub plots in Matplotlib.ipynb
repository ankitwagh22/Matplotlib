# Introduction to Figures and sub plots in Matplotlib

```python
import matplotlib.pyplot as plt

# changing the figure size
plt.figure(figsize=(10,6))

x = [1,2,3,4,5,6]
y = [2,1,5,4,7,5]

plt.title("Changing the figure size")
plt.plot(x,y,marker='d',ms=10,mfc='r')
plt.show()
```

```python
# creating subplotes
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,6))

# adding the sub-plot to created figure
fig1 = fig.add_subplot(111)

x = [1,2,3,6,7,8]
y = [3,4,6,1,3,0] 

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Adding the sub plot")

# plotting points on subplot
fig1.plot(x,y)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()

# adding the sub plot
fig1 = fig.add_subplot(1,2,1)
lst1 = [1,2,5,7]
fig1.plot(lst1,marker='o')

# subplot -2
fig2 = fig.add_subplot(1,2,2)
lst2 = [1,2,5,0]
fig2.plot(lst2,marker='d')

# add title of subplots 
fig.suptitle("Figure with multiple subplot")
plt.show()
```

```python
# Example: Multiple plots on the same figure

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,6))

# subplot-1
fig1 = fig.add_subplot(2,2,1)
fig1.plot([1,2,5,7])

# subplot-2
fig2 = fig.add_subplot(2,2,2)
fig2.plot([1,2,5,0])

# subplot-3
fig3 = fig.add_subplot(2,2,3)
fig3.plot([3,2,7,0])

# subplot-4
fig4 = fig.add_subplot(2,2,4)
fig4.plot([0,8,8,0])

fig.suptitle('Multiple subplots')
plt.show()
```

```python
# Drawing Multiple Sub Plots with Figure Object

import matplotlib.pyplot as plt
import numpy as np

# plot-1
arr1 = np.array([1,2,3,4])
arr2 = np.array([2,3,4,5])
plt.subplot(1,2,1)
plt.plot(arr1,arr2)

# plot-2
arr3 = np.array([1,3,5,7])
arr4 = np.array([6,8,10,12])
plt.subplot(1,2,2)
plt.plot(arr3,arr4)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

# plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```
