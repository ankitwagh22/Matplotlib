# Line plot examples using pandas and matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'Days': ['Monday','Tuesday','Wednesday','Thursday','Friday', 'Saturday', 'Sunday'], 
        'Week 1': [27,29,32,33,33,31,29],
        'Week 2': [29,30,31,33,34,34,31],
        'Week 3': [29,28,25,27,31,30,29]}

df = pd.DataFrame(data)
print(df)
```

```python
df['Days']
```

```python
df['Week 1']
```

```python
df[["Week 1","Week 2","Week 3"]]
```

```python
print(df)
```

```python
# ploting for week 1
plt.plot(df['Days'],df['Week 1'],c='red',marker='o',mfc='b')
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("First week Temperature")
plt.show()
```

```python
# ploting for week -2
plt.plot(df['Days'],df['Week 2'],c='red',mfc='b',marker='o')
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Second week Temperature")
plt.show()
```

```python
# ploting for week -3
plt.plot(df['Days'],df['Week 3'],c='red',mfc='blue',marker='D')
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Third week Temperature")
plt.show()
```

```python
# ploting all the weeks temperature at same graph
plt.plot(df['Days'],df['Week 1'],c='red',marker='o',mfc='yellow')
plt.plot(df['Days'],df['Week 2'],c='green',marker='D',mfc='pink')
plt.plot(df['Days'],df['Week 3'],c='orange',marker='d',mfc='blue')

f1 = {'font':'Arial','color':'blue','size':15}
plt.xlabel("Days",fontdict=f1)
plt.ylabel("Temperature",fontdict=f1)
plt.title("Weekly Temperature Analysis",fontdict=f1)
lab = ['Week-1','Week-2','Week-3']
plt.legend(labels=lab,loc='best')
plt.show()
```

```python
# saving the figure on our disk
plt.plot(df['Days'],df['Week 1'],c='red',marker='o',mfc='yellow')
plt.plot(df['Days'],df['Week 2'],c='green',marker='D',mfc='pink')
plt.plot(df['Days'],df['Week 3'],c='orange',marker='d',mfc='blue')

f1 = {'font':'Arial','color':'blue','size':15}
plt.xlabel("Days",fontdict=f1)
plt.ylabel("Temperature",fontdict=f1)
plt.title("Weekly Temperature Analysis",fontdict=f1)
lab = ['Week-1','Week-2','Week-3']
plt.legend(labels=lab,loc='best')
plt.savefig("D:\\Full stack Python Notes\\temp.jpeg")
plt.show()
```

```python
# Case: Comparison of Product Sales for 12 Months
# create synthetic data for product sales
month = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
product_sales_A = np.random.randint(50,200, 12) * 100
product_sales_B = np.random.randint(50,200, 12) * 100

df_sales = pd.DataFrame ({ 'month' : month,
						'Product A': product_sales_A,
						'Product B': product_sales_B })
print(df_sales)
```

```python
plt.plot(df_sales['month'],df_sales['Product A'],c='red',marker='d',mfc='yellow')
plt.show()
```

```python
plt.plot(df_sales['month'],df_sales['Product B'],c='blue',linestyle='dotted',marker='o',mfc='red')
plt.show()
```

```python
# combining both the graphs
plt.plot(df_sales['month'],df_sales['Product A'],c='red',marker='D',mfc='blue')
plt.plot(df_sales['month'],df_sales['Product B'],c='blue',marker='o',mfc='yellow',linestyle='dotted')

plt.xlabel('Months')
plt.ylabel("Sales")
plt.title("Products and their sales analysis")
plt.legend(labels = ['Product-A','Product-B'],loc='best')
plt.show()
```

```python
# The table represents Akshays weight in kilograms for five months. 
# Draw the line graph for the information provided Month
# Akshays weight in kilogram

import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("E:\\Askshay.csv")
print(data)
```

```python
data.loc[0:1].values
```

```python
month = data.columns.values
weight = data.iloc[0].values
plt.plot(month,weight,c='blue',marker='o',mfc='red',ms=8)
plt.xlabel('Months')
plt.ylabel('Akshay weight')
plt.title('Monthly akshay weight anaysis')
plt.show()
```
