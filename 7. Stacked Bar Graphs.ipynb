# Stacked Bar Graphs

```python
# years		CSE			IT		ECE
# ----------	     ------------         
# 2019		70			30		20
# 2020		40			20		19
# 2021		60			25		20
# 2022		80			35		18

#Display the above Placement of Certain College in the Form of Stacked Bar Graph
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
years = [2019,2020,2021,2022]
cse = [70,40,60,80]
it = [30,20,25,35]
ece = [20,19,20,18]

plt.xlabel('Years')
plt.ylabel('No. of students')
plt.title('Placement Ratio Data')
plt.bar(years,cse,color='purple',width=0.5)
plt.show()
```

```python
#Draw the Bar graph year and cse
years = [2019,2020,2021,2022]
cse = [70,40,60,80]
it = [30,20,25,35]
ece = [20,19,20,18]

plt.bar(years,cse)
plt.bar(years,it,bottom=cse)
plt.show()
```

```python
years = [2019,2020,2021,2022]
cse = [70,40,60,80]
it = [30,20,25,35]
ece = [20,19,20,18]
ece_start = [cse[i]+it[i] for i in range(4)]

plt.bar(years,cse)
plt.bar(years,it,bottom=cse)
plt.bar(years,ece,bottom=ece_start)
plt.show()
```

```python
years = [2019,2020,2021,2022]
cse = [70,40,60,80]
it = [30,20,25,35]
ece = [20,19,20,18]
ece_start = [cse[i]+it[i] for i in range(4)]

plt.bar(years,cse,color='green')
plt.bar(years,it,bottom=cse,color='purple')
plt.bar(years,ece,bottom=ece_start,color='hotpink')
plt.show()
```
