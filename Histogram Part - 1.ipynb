# Histogram Part - 1 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
StudentMarks = [90,50,40,60,55,44,30,10,34,84]
MarksInterval = [0,35,75,100]
plt.hist(StudentMarks,MarksInterval)
plt.xlabel('Students marks')
plt.ylabel('Marks Interval')
plt.show()
```

```python
StudentMarks = [90,50,40,60,55,44,30,10,34,84]
MarksInterval = [0,35,75,100]
plt.hist(StudentMarks,MarksInterval,edgecolor='yellow')
plt.xticks([0,35,75,100])
plt.xlabel('Students marks')
plt.ylabel('Marks Interval')
plt.show()
```

```python
StudentMarks = [90,50,40,60,55,44,30,10,34,84]
MarksInterval = [0,35,75,100]
plt.hist(StudentMarks,MarksInterval,histtype='bar',edgecolor='yellow')
plt.xticks([0,35,75,100])
plt.xlabel('Students marks')
plt.ylabel('Marks Interval')
plt.show()
```

```python
StudentMarks = [90,50,40,60,55,44,30,10,34,84]
MarksInterval = [0,35,75,100]
plt.hist(StudentMarks,MarksInterval,histtype='step',edgecolor='red')
plt.xticks([0,35,75,100])
plt.xlabel('Students marks')
plt.ylabel('Marks Interval')
plt.show()
```

```python
StudentMarks = [90,50,40,60,55,44,30,10,34,84]
MarksInterval = [0,35,75,100]
plt.hist(StudentMarks,MarksInterval,histtype='stepfilled',edgecolor='yellow')
plt.xticks([0,35,75,100])
plt.xlabel('Students marks')
plt.ylabel('Marks Interval')
plt.show()
```

```python
StudentMarks = [90,50,40,60,55,44,30,10,34,84]
MarksInterval = [0,35,75,100]
plt.hist(StudentMarks,MarksInterval,histtype='bar',edgecolor='yellow',rwidth=0.5)
plt.xticks([0,35,75,100])
plt.xlabel('Students marks')
plt.ylabel('Marks Interval')
plt.show()
```
