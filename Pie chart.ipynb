# Pie Chart 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.array([10,5,5,10,10])
plt.pie(data)
plt.show()
```

```python
# Label attrinute

data = np.array([10,5,5,10,10])
games = np.array(["FootBall","Hockey","Cricket","BasketBall","Badminton"])
plt.pie(data,labels=games)
plt.show()
```

```python
# Attrinute -> startangle

plt.pie(data,labels=games,startangle=90)
plt.show()
```

```python
# Attribute --> explode

plt.pie(data,labels=games,startangle=0,explode=[0.3,0,0,0,0])
plt.show()
```

```python
# mutiple explode

plt.pie(data,labels=games,startangle=0,explode=[0.1,0,0.3,0,0])
plt.show()
```

```python
# argument --> shadow

plt.pie(data,labels=games,startangle=90,explode=[0.2,0,0.3,0,0],shadow=True)
plt.show()
```

```python
# argument --> color

plt.pie(data,labels=games,startangle=0,explode=[0.2,0,0.3,0,0],colors=['red','yellow','green','hotpink','purple'])
plt.title("This is Pie chart")
plt.show()
```

```python
# We can also gives the legend to our diagram

plt.pie(data,labels=games,startangle=0,explode=[0.2,0,0.3,0,0])
plt.legend(title='Games')
plt.show()
```

```python
# argument ----> autopct=%0.0f%%

plt.pie(data,labels=games,autopct='%0.0f%%',explode=[0.2,0,0,0,0])
plt.show()
```

```python
plt.pie(data,labels=games,autopct='%0.2f%%',explode=[0.2,0,0,0,0])
plt.show()
```

```python
# Argument --> wedgeprops

plt.pie(data,labels=games,explode=[0,0,0.2,0,0],wedgeprops={'linewidth':2,'edgecolor':'red'})
plt.show()
```

```python
plt.pie(data,labels=games,startangle=0,explode=[0.2,0,0.3,0,0],colors=['cyan','yellow','green','hotpink','purple'],
       wedgeprops={'linewidth':2,'edgecolor':'red'})
plt.title("This is Pie chart")
plt.show()
```

```python
# Question: Priya lists down her monthly expenditure as follows:

#	Expenditure	Amount
#			Rent			4000
#			Food		5400
#			Clothing		2800
#			Savings		400
#		Draw a pie chart for her monthly expenses.
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dic = {'Rent':4000,'Food':5400,'Clothing':2800,'Saving':400}
df = pd.DataFrame(dic,index=[0])
print(df)
```

```python
df.loc[0].values
```

```python
df.columns
```

```python
# Drawing the pie chart for that
plt.pie(df.loc[0].values)
plt.title('Priya Expenses')
plt.show()
```

```python
plt.pie(df.loc[0].values,labels=df.columns)
plt.title('Priya Expenses')
plt.show()
```

```python
plt.pie(df.loc[0].values,labels=df.columns,explode=[0,0.1,0,0])
plt.title('Priya Expenses')
plt.show()
```

```python
plt.pie(df.loc[0].values,labels=df.columns,explode=[0,0.1,0,0],autopct='%0.0f%%')
plt.title('Priya Expenses')
plt.show()
```

```python
cls = ['yellow','blue','hotpink','purple']
plt.pie(df.loc[0].values,labels=df.columns,explode=[0,0.1,0,0],autopct='%0.0f%%',colors=cls)
plt.title('Priya Expenses')
plt.show()
```

```python
cls = ['yellow','blue','hotpink','purple']
plt.pie(df.loc[0].values,labels=df.columns,explode=[0,0.1,0,0.5],autopct='%0.0f%%',colors=cls,
       wedgeprops={'linewidth':3,'edgecolor':'green'})
plt.title('Priya Expenses')
plt.show()
```
